# 🧱 Step 1 — Proper Data Type Correction

```python
consolidated_df['event_time'] = pd.to_datetime(
    consolidated_df['event_time'],
    utc=True,
    errors='coerce'
)
```

```python
consolidated_df['event_time'].isna().sum()
```

# 🧹 Step 2 — Clean Critical Fields

## 1. Remove exact duplicates

```python
consolidated_df = consolidated_df.drop_duplicates()
```

## 2. Handle missing sessions

```python
consolidated_df = consolidated_df.dropna(subset=['user_session'])
```

# 🧠 Step 3 — Feature Engineering

## A) Revenue Column

```python
consolidated_df['revenue'] = np.where(
    consolidated_df['event_type'] == 'purchase',
    consolidated_df['price'],
    0
)
```

## B) Date Features

```python
consolidated_df['date'] = consolidated_df['event_time'].dt.date
consolidated_df['hour'] = consolidated_df['event_time'].dt.hour
consolidated_df['day_of_week'] = consolidated_df['event_time'].dt.weekday
```

# 🏗 Step 4 — Create Curated Analytical Tables

## 1️⃣ Session-Level Fact Table

```python
session_fact = consolidated_df.groupby('user_session').agg(
    user_id=('user_id', 'first'),
    category_id=('category_id', 'first'),
    session_start=('event_time', 'min'),
    session_end=('event_time', 'max'),
    total_events=('event_type', 'count'),
    total_views=('event_type', lambda x: (x == 'view').sum()),
    total_carts=('event_type', lambda x: (x == 'cart').sum()),
    total_purchases=('event_type', lambda x: (x == 'purchase').sum()),
    revenue=('revenue', 'sum')
).reset_index()

session_fact['session_duration_minutes'] = (
    (session_fact['session_end'] - session_fact['session_start'])
    .dt.total_seconds() / 60
)
```

## 2️⃣ Product-Level Fact Table

```python
product_fact = (
    consolidated_df
    .groupby(['date', 'product_id'])
    .agg(
        category_id=('category_id', 'first'),
        brand=('brand', 'first'),
        avg_price=('price', 'mean'),
        total_views=('event_type', lambda x: x.eq('view').sum()),
        total_carts=('event_type', lambda x: x.eq('cart').sum()),
        total_purchases=('event_type', lambda x: x.eq('purchase').sum()),
        revenue=('revenue', 'sum')
    )
    .reset_index()
)
```

## 3️⃣ Daily Metrics Table

```python
daily_metrics = consolidated_df.groupby('date').agg(
    total_events=('event_type', 'count'),
    total_views=('event_type', lambda x: (x == 'view').sum()),
    total_carts=('event_type', lambda x: (x == 'cart').sum()),
    total_purchases=('event_type', lambda x: (x == 'purchase').sum()),
    revenue=('revenue', 'sum')
).reset_index()
```

# 🧪 Step 5 — Strategic Quality Check
Check session duration sanity

```python
session_fact['session_duration_minutes'].describe(percentiles=[0.95,0.99])
```

# 💾 Step 6 — Export Curated Layer

```python
session_fact.to_csv("session_fact.csv", index=False)
product_fact.to_csv("product_fact.csv", index=False)
daily_metrics.to_csv("daily_metrics.csv", index=False)
```
